```python
from surprise import Dataset, SVD
from surprise.model_selection import cross_validate

# 1. Load a built-in dataset (no CSV needed).
# This will download MovieLens 100k the first time you run it.
data = Dataset.load_builtin('ml-100k')

# 2. Build a full trainset (use all ratings for training).
trainset = data.build_full_trainset()

print(type(data))
print(type(trainset))

# 3. Create the SVD model with 2 latent factors (like your example).
model = SVD(n_factors=2)

# 4. Fit the model on the full trainset.
model.fit(trainset)

# 5. Build a testset from the trainset (all known ratings),
#    then get a list of prediction objects.
testset = trainset.build_testset()
predictions_list = model.test(testset)

# 6. Run cross-validation on the original Dataset object
#    (this time with more factors so the model has some capacity).
cross_val_results = cross_validate(
    model,
    data,
    measures=['RMSE'],
    cv=3,
    verbose=True
)

# ANSWER CHECK
print(cross_val_results)
```
