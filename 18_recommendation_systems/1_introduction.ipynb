A recommendation system is an ML-based system that predicts how much a user will like an item (often as a rating or relevance score) and then ranks items accordingly to show the most promising ones first. It is widely used for products, movies, music, news, and more to help users navigate large catalogs efficiently. [learnopencv](https://learnopencv.com/recommendation-system/)

## What a recommendation system does

- It collects user–item interaction data such as ratings, clicks, views, likes, and purchases, plus item metadata like category, text, or tags. [geeksforgeeks](https://www.geeksforgeeks.org/machine-learning/what-are-recommender-systems/)
- It learns patterns that link users to items they tend to prefer, then predicts a score for unseen items and recommends the top-ranked ones. [nvidia](https://www.nvidia.com/en-us/glossary/recommendation-system/)
- Modern systems usually operate in stages: candidate generation (narrowing the catalog), scoring (assigning relevance), and re-ranking (adjusting for diversity, freshness, or business rules). [developers.google](https://developers.google.com/machine-learning/recommendation/overview/types)

A simple example is Netflix predicting how much you will like each movie, assigning a score, and then showing you a “Top Picks for You” row ordered by those scores. [learnopencv](https://learnopencv.com/recommendation-system/)

## Main types of recommendation systems

- **Content-based filtering**: Recommends items similar to those the user liked in the past, based on item features (e.g., movie genre, keywords, textual description, or product attributes). [d2l](http://d2l.ai/chapter_recommender-systems/recsys-intro.html)
- **Collaborative filtering**: Uses patterns in the user–item interaction matrix (who liked what) to recommend items that similar users liked or items that tend to be co-liked, without needing rich item metadata. [geeksforgeeks](https://www.geeksforgeeks.org/machine-learning/what-are-recommender-systems/)
- **Hybrid systems**: Combine collaborative and content-based (and sometimes contextual) signals to get better accuracy, handle cold-start cases, and improve robustness. [itransition](https://www.itransition.com/machine-learning/recommendation-systems)

A hybrid video recommender might use user–video watch history (collaborative) plus video tags and text embeddings (content-based) to decide which videos to show on the home page. [hellointerview](https://www.hellointerview.com/learn/ml-system-design/problem-breakdowns/video-recommendations)

A recommendation engine is an ML system that predicts user ratings for items and returns ranked results, with three main families: content-based, collaborative, and hybrid.
