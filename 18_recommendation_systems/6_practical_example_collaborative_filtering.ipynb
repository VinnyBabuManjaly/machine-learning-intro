```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'Alfred': [3.0, 4.0, np.nan, 4.0, 4.0],
    'Mandy': [np.nan, 9.0, np.nan, 3.0, 8.0],
    'Lenny': [2.0, 5.0, 8.0, 9.0, np.nan],
    'Joan': [3.0, np.nan, 9.0, 4.0, 9.0],
    'Tino': [1.0, 1.0, np.nan, 9.0, 5.0]
}

reviews = pd.DataFrame(data, index=['Michael Jackson', 'Clint Black', 'Dropdead', 'Anti-Cimex', 'Cardi B'])

reviews
```

```python
# Creating F1 and F2
# To begin, create two randomly instantiated vectors F1 and F2 as columns in your DataFrame.

np.random.seed(42)
reviews['F1'] = np.random.normal(size = 5)
reviews['F2'] = np.random.normal(size = 5)

reviews
```

```python
# Regression models for all users
uf = []
for c in reviews.columns[:5]:
    X = reviews.dropna(subset = [c])[['F1', 'F2']]
    y = reviews[c].dropna()
    lr = LinearRegression(fit_intercept=False).fit(X, y)
    coefs = lr.coef_
    uf.append(list(coefs))
uf = np.array(uf)

uf.shape
```

```python
# New Model for artists
ui_df = reviews.iloc[:, :-2].T
ui_df['F1'] = uf[:, 0]
ui_df['F2'] = uf[:, 1]
ui_df
```

```python
# Regression models for all artists
ifs = []
for c in ui_df.columns[:5]:
    X = ui_df.dropna(subset = [c])[['F1', 'F2']]
    y = ui_df[c].dropna()
    lr = LinearRegression(fit_intercept=False).fit(X, y)
    coefs = lr.coef_
    ifs.append(list(coefs))
ifs = np.array(ifs)
```

```python
if_df = reviews.copy().iloc[:, :-2]
if_df.loc[:, 'F1'] = ifs[:, 0]
if_df.loc[:, 'F2'] = ifs[:, 1]
if_df
```

```python
uf2 = []
for c in if_df.columns[:5]:
    X = if_df.dropna(subset = [c])[['F1', 'F2']]
    y = if_df[c].dropna()
    lr = LinearRegression(fit_intercept=False).fit(X, y)
    coefs = lr.coef_
    uf2.append(list(coefs))
uf2 = np.array(uf2)
```

```python
ui_df2 = reviews.copy().iloc[:, :-2].T
ui_df2['F1'] = uf2[:, 0]
ui_df2['F2'] = uf2[:, 1]
ui_df2
```

```python
ifs2 = []
for c in ui_df2.columns[:5]:
    X = ui_df2.dropna(subset = [c])[['F1', 'F2']]
    y = ui_df2[c].dropna()
    lr = LinearRegression(fit_intercept=False).fit(X, y)
    coefs = lr.coef_
    ifs2.append(list(coefs))
ifs2 = np.array(ifs2)
```

```python
if_df2 = reviews.copy().iloc[:, :-2]
if_df2.loc[:, 'F1'] = ifs2[:, 0]
if_df2.loc[:, 'F2'] = ifs2[:, 1]
if_df2
```

```python
if_df.to_csv('data/Q.csv')
ui_df.to_csv('data/P.csv')
```
