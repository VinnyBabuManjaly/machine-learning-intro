# pandas Operations on DataFrames

### What Is a Series?
A *Series* is a single column in a pandas DataFrame, acting like a one-dimensional labeled array. 

DataFrames are two-dimensional structures made up of multiple Series.

Series have many built-in methods like *.sum(), .multiply(), .add()*, and more to perform vectorized (element-wise) operations efficiently.


### Basic Arithmetic Operations on Columns
Operations on pandas Series (columns) resemble spreadsheet calculations and can be performed either via operators or methods.

These operations are efficient and behave like vectorized array operations in NumPy but preserve labels and indexing.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [0.1, 0.2, 0.3],
    'B': [4, 5, 6]
})
df
```

```python
# Summation of a column

sumA = df['A'].sum()
print(sumA)
```

```python
# Addition of two columns using add method

df['C'] = df['A'].add(df['B'])
df
```

```python
# Addition of two columns using operator

df['D'] = df['B'] + df['C']
df
```

```python
# Multplication of two colums using multiply method

df['E'] = df['C'].multiply(df['D'])
df
```

```python
# Calculating percentage of column by dividing to its sum

df['PercentE'] = df['E'] / df['E'].sum() * 100
df
```

### Plotting with pandas
pandas integrates basic plotting capabilities via Matplotlib for quick visualizations.

**Basic Plot Syntax**:  

    df.plot(kind='plot_kind')

The *kind* keyword argument specifies the plot type, e.g., 'line', 'bar', 'hist', 'scatter', etc.

```python
# Example Dataset: Iris Flowers (from scikit-learn)

from sklearn.datasets import load_iris

iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target

print(df_iris.head())

# Data contains sepal length, sepal width, petal length, petal width, and species target.
```

```python
# Histogram

# All numerical columns:
df_iris.plot(kind='hist', alpha=0.5, bins=20)  # alpha to control transparency

# # To save histogram or any plot as pdf file
# histPlot = df_iris.plot(kind='hist')
# histPlot.figure.savefig('plots/hist-plot.pdf')

# Selected columns only, e.g., sepal length and width:
df_iris.plot(kind='hist', y=['sepal length (cm)', 'sepal width (cm)'], bins=20, alpha=0.7)
```

```python
# Scatter Plot
# Visualize relationship between two variables:

df_iris.plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)', grid=True)
grid=True # adds grid lines to the plot.

# Scatter plots visualize correlations and trends.
```

```python
# Line plot

# Timeseries dataframe for line plot
df_time_series = pd.read_csv('timeseries.csv')
# Sample line plot pseudo code
df_time_series.plot(kind='line', x='Date', y='Consumption', figsize=(12,6), grid=True)

# Multiple lines can be plotted by providing a list for the y argument:
df_time_series.plot(kind='line', x='Date', y=['Wind', 'Solar'], figsize=(12,6), grid=True)

# figsize controls size of plot.
# Good for trend analysis over time.
```

```python
# Bar Plot for Summary Aggregates

totals = df_time_series[['Consumption', 'Wind', 'Solar']].sum()
totals.plot(kind='bar')
```

- pandas plotting builds on Matplotlib, so plot objects can be further customized.

- For advanced plotting needs, libraries like Seaborn or Plotly are recommended.

- Vectorized operations in pandas Series and DataFrames allow efficient data processing and clean code.
