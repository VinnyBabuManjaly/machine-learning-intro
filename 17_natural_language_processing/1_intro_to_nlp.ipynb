## Natural Language Processing (NLP)

### Overview of NLP

Natural Language Processing (NLP) is a branch of artificial intelligence that focuses on enabling computers to interpret, understand, and generate human language. It has two main subfields:

- **Natural Language Understanding (NLU):**  
  Involves analyzing text to derive meaning and intent based on its syntax and semantics.  
  *Example:* Detecting whether a review is positive or negative by understanding sentence structure and word meanings.

- **Natural Language Generation (NLG):**  
  Involves producing natural-sounding language from structured data (and sometimes adding speech synthesis).  
  *Example:* ChatGPT generating text replies or automated news writing.

### Introduction to NLP

Why NLP is challenging and distinct from traditional, number-based machine learning:

1. **Word Order Matters:**  
   The sequence of words affects meaning (e.g., *“I cleaned my car”* vs. *“I had my car cleaned”*).

2. **Word Meaning Depends on Context:**  
   Words like *“bark”* change meaning depending on the sentence.

3. **Human Language Is Imprecise:**  
   Language includes ambiguity, idioms, typos, and personalization.

**Applications** include:
- Search engines (query understanding)
- Spam detection
- Machine translation (Google Translate)
- Dialogue systems and chatbots
- Text summarization and categorization
- Sentiment analysis (e.g., movie or tweet opinions)
- Speech recognition (Alexa, Siri, Google Assistant)

**Workflow comparison with ML:**
- Data in NLP is called a *corpus*, and each sample is a *document* (review, article, tweet, etc.). 

- Similar ML workflow applies—split data, preprocess, extract features, train, and evaluate models.

- Key NLP-specific steps:

  - **Tokenization:** Split text into words or units.
  - **Normalization:** Convert tokens to a standard form (e.g., lowercase, removing punctuation).
  - **Feature extraction:** Convert text into numbers using approaches like *bag-of-words* or *TF‑IDF*.

### Mini‑Lesson 18.1: Applications of NLP
NLP powers tasks across industries. Major application areas include:

- **Text Classification & Sentiment Analysis:** Spam detection, customer review analysis.
- **Named Entity Recognition (NER):** Identifying names, locations, or organizations in text.
- **Machine Translation:** Cross-language translation (e.g., Google Translate).
- **Speech Recognition & Voice Assistants:** Converting speech ↔ text for systems like Alexa or Siri.
- **Text Generation:** Producing meaningful text from prompts (e.g., chatbots, storytelling apps).
- **Question Answering:** Generating answers to natural language queries.
- **Summarization:** Extracting key points from long texts.
- **Document Understanding:** Structuring and categorizing information from documents.
- **Social Media Analysis:** Identifying trends, fake news detection, and brand monitoring.
- **Healthcare NLP:** Extracting insights from medical texts for diagnosis or drug discovery.
- **Legal, Education, Finance, and Customer Support:** Contract analysis, tutoring systems, market sentiment analysis, and automated support.
