## Why Preprocess Text?
Raw text has "noise" like emotions (!!!), punctuation, mixed case (Apple vs apple), and different ways to say the same thing ("happy" vs "joyful"). Computers can't read words directly, so preprocessing cleans it up and converts to numbers for ML models. It also shrinks data size (reduces dimensions) to make analysis faster and clustering better. [mbrenndoerfer](https://mbrenndoerfer.com/writing/text-preprocessing-nlp-tokenization-normalization)

## Overview: Key Preprocessing Steps

These steps prepare text for models:

- **Tokenization:** Break text into small units (tokens like words or sentences). 

- **Lower casing:** Make all text lowercase to ignore case differences.

- **Stop words removal:** Drop common words like "the," "is," "and" that don't add much meaning.

- **Stemming:** Chop words to root form (e.g., "running" → "run").

- **Lemmatization:** Smarter stemming using dictionary (e.g., "better" → "good"). [dev](https://dev.to/themustaphatijani/the-complete-guide-to-nlp-text-preprocessing-tokenization-normalization-stemming-lemmatization-50ap)
