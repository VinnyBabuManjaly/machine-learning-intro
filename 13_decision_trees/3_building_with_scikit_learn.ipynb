Decision trees in scikit-learn fit just like other estimators: create a `DecisionTreeClassifier()`, call `.fit(X, y)` on training data, and use `.predict()` for class predictions on new observations.

## Basic Python Implementation
Load the Iris dataset, split data, and train:

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25, random_state=42)
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)

# This mirrors the process on 150 Iris samples for Setosa, Versicolor, and Virginica classification.
```

## Visualizing the Tree
Scikit-learn's `plot_tree` generates diagrams: `plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True, filled=True)`. 

Default layout has unclear arrows, so use `export_graphviz` for better Graphviz output:

```python
from sklearn.tree import export_graphviz
import graphviz
dot_data = export_graphviz(clf, out_file=None, feature_names=iris.feature_names, 
                          class_names=iris.target_names, filled=True, rounded=True)
graph = graphviz.Source(dot_data)
graph
```

## Model Comparison
Decision tree boundaries appear nonlinear and stepwise (100% training accuracy by eye), contrasting logistic regression's straight lines on Iris petal plots, with accuracy computation next.
