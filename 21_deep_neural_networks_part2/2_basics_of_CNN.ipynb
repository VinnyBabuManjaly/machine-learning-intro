CNNs are neural networks that are especially good at understanding images and videos. They automatically learn useful visual patterns (like edges, shapes, and objects) directly from raw pixels, without you hand‑coding features.



## What CNNs are used for

- **Image classification**: Decide which category an image belongs to (for example, “cat”, “dog”, “car”).  
- **Object detection**: Find and label multiple objects in an image and mark where they are.  
- **Image segmentation**: Color or “cut out” different regions of an image so each object or part has its own region.  
- **Face recognition**: Recognize who a person is from their face.  
- **Image generation**: Create new realistic images after learning from many examples.  
- **Style transfer**: Make one image look like it was drawn or painted in the style of another.  
- **Visual question answering**: Look at an image and answer questions about what is in it.



## Main building blocks of a CNN

### 1. Input layer

- This is where you feed the image into the network.  
- An image is stored as numbers in a grid: height × width × channels (for example, RGB has 3 channels).  



### 2. Convolutional layer

- This is the **core** layer in a CNN.  
- It uses many small matrices called **filters** (or **kernels**) that slide over the image.  
- At each position, the filter and the underlying pixels are multiplied and summed to produce one number. Repeating this over all positions builds a **feature map** that shows where a certain pattern appears.

Key ideas inside a convolutional layer:

- **Filters / kernels**:  
  - Small trainable weight matrices.  
  - Each filter specializes in detecting some pattern (for example, edges or textures).  

- **Stride**:  
  - How many pixels you move the filter each step (1 = move by 1 pixel, 2 = skip every other pixel, and so on).  
  - Larger stride → smaller output, less detail, faster computation.  

- **Padding**:  
  - Adding extra pixels (often zeros) around the image border.  
  - Used to control the size of the output (for example, to keep width/height the same).  



### 3. Activation function

- Applied after the convolution to add **nonlinearity**.  
- The most common one in CNNs is **ReLU**, which turns negative values into 0 and keeps positive values unchanged.  
- This lets the network learn complex patterns, not just simple linear relationships.



### 4. Pooling layer

- Also called **subsampling** or **downsampling**.  
- Its job is to reduce width and height of feature maps while keeping the most important information.  
- The most common type is **max pooling**, which takes the largest value in a small region.  

Why pooling helps:

- Reduces computation and memory.  
- Makes features more robust to small shifts or changes in the image.  
- Forces the network to focus on “strongest” signals, not tiny details.



### 5. Flattening

- After several convolution + activation + pooling layers, you have several small feature maps.  
- **Flattening** converts this 3D block (height × width × channels) into a 1D vector of numbers.  
- This prepares the data to feed into fully connected layers.



### 6. Fully connected layers

- These are standard dense layers: each neuron connects to all outputs from the previous layer.  
- They combine all the extracted features to make a decision.  
- Here, the network learns how different features relate to the final task (for example, which combination indicates a cat vs. a dog).



### 7. Output layer

- The final layer produces the **result** of the model.  
- In classification, this is typically a vector of **class scores**, passed through **softmax** to get probabilities for each class (for example, digits 0–9).



## How CNNs learn (training)

- CNNs start with random weights in filters and fully connected layers.  
- For each image, the network makes a prediction and compares it to the true label using a **loss function** (a measure of error).  
- Using **backpropagation**, the model computes how every weight contributed to the error.  
- An **optimizer** (like SGD or Adam) adjusts the weights slightly to reduce the loss.  
- Repeating this over many images and epochs gradually shapes the filters to detect useful features and improves accuracy.



## After training: evaluation and inference

- **Evaluation**: Test the trained CNN on new, unseen data to measure how well it generalizes.  
- **Inference**: Use the trained CNN to make predictions on real‑world input (new images or videos).  



## Big picture

A CNN works by stacking these operations:

1. Convolution + activation → detect local patterns.  
2. Pooling → make patterns more robust and reduce size.  
3. Repeat steps 1–2 to build a feature hierarchy.  
4. Flatten + fully connected + output → use learned features to make the final prediction.  

This pipeline lets CNNs turn raw pixels into meaningful, high‑level understanding of visual data.
