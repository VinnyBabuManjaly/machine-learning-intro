A convolutional neural network (CNN) is a special kind of neural network that is very good at understanding images (and other grid‑like data such as audio spectrograms). [towardsdatascience](https://towardsdatascience.com/convolutional-neural-networks-for-beginners-c1de55eee2b2/)



### Why plain dense networks are a bad fit for images

- In a dense (fully connected) network, you first **flatten** an image into one long list of pixels, then connect every pixel to every neuron in the next layer. [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- Flattening destroys the original 2D layout of the image, so the network has to “relearn” where things are and how pixels relate in space. [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- Natural images have structure: nearby pixels tend to be similar (spatial correlation), objects are built hierarchically from simple parts, and the object type does not change if you move it around in the image (translation invariance). [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- Dense layers do not take advantage of any of this structure; they treat every pixel as unrelated to its neighbors. [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

CNNs are designed specifically to use these image regularities rather than ignore them. [towardsdatascience](https://towardsdatascience.com/convolutional-neural-networks-for-beginners-c1de55eee2b2/)



### Key idea: receptive field

- A **receptive field** is the small region of the input that a given neuron “looks at.” [stanford](https://stanford.edu/~shervine/teaching/cs-230/cheatsheet-convolutional-neural-networks)

- In CNNs, neurons in early layers see tiny patches of the image; neurons in deeper layers see larger, more complex regions built from earlier features. [stanford](https://stanford.edu/~shervine/teaching/cs-230/cheatsheet-convolutional-neural-networks)

- This local view lets the network detect simple patterns (like edges) and then gradually combine them into more complex shapes. [ibm](https://www.ibm.com/think/topics/convolutional-neural-networks)



### Core operation: convolution

- A **convolution** uses a small grid of numbers called a **kernel** or **filter** that slides over the image. [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- At each position, the kernel is multiplied with the underlying pixel values and summed to produce one number; this is then adjusted by a **bias** and passed through an **activation function** such as ReLU. [kitchell.github](https://kitchell.github.io/DeepLearningTutorial/4cnnsinkeras.html)

- The kernel, bias, and activation together form a **filter**; each filter is trained to respond strongly to a particular visual pattern (for example, a certain kind of edge). [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- Sliding the same filter over all positions creates a 2D output called a **feature map**, which shows where that pattern appears in the image. [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- The filter weights and bias are not hand‑designed; they are learned during training. [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)



### Padding and stride

- **Padding** adds extra pixels (often zeros) around the border so that the filter can cover edge pixels without shrinking the output. [machinelearningmastery](https://www.machinelearningmastery.com/crash-course-convolutional-neural-networks/)

  - With no padding (often called “valid”), the output is smaller than the input.  
  - With padding chosen carefully (often called “same”), the output has the same height and width as the input. [kitchell.github](https://kitchell.github.io/DeepLearningTutorial/4cnnsinkeras.html)

- **Stride** controls how far the filter moves each time. [machinelearningmastery](https://www.machinelearningmastery.com/crash-course-convolutional-neural-networks/)

  - A stride of 1 means the filter moves one pixel at a time, giving a large, detailed feature map. [machinelearningmastery](https://www.machinelearningmastery.com/crash-course-convolutional-neural-networks/)

  - A larger stride skips positions, making the output smaller and reducing computation. [machinelearningmastery](https://www.machinelearningmastery.com/crash-course-convolutional-neural-networks/)



### Convolutional layers

- A **convolutional layer** uses many filters in parallel, all with the same kernel size, stride, padding, and activation, but different learned weights. [kitchell.github](https://kitchell.github.io/DeepLearningTutorial/4cnnsinkeras.html)

- Each filter produces its own feature map; stacking these feature maps forms the output of that layer. [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- Early convolutional layers learn low‑level patterns; deeper layers learn more abstract features by combining earlier ones. [ibm](https://www.ibm.com/think/topics/convolutional-neural-networks)



### Pooling layers (especially max pooling)

- A **pooling layer** shrinks feature maps while keeping the most important information. [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- **Max pooling** divides the feature map into small regions and outputs the maximum value from each region. [geeksforgeeks](https://www.geeksforgeeks.org/deep-learning/cnn-introduction-to-pooling-layer/)

- Pooling:  
  - Reduces spatial size, cutting memory and computation.  
  - Makes features more robust to small shifts and distortions in the input.  
  - Helps build a hierarchy: after pooling, new convolutions see a “zoomed out” view and can detect higher‑level features. [machinelearningmastery](https://www.machinelearningmastery.com/crash-course-convolutional-neural-networks/)



### From 2D to 3D kernels

- After the first convolutional layer, the data is no longer a single 2D image but a stack of feature maps (a 3D block: height × width × channels). [ibm](https://www.ibm.com/think/topics/convolutional-neural-networks)

- Filters in deeper layers are therefore **3D kernels**: they span a small spatial region (for example, 3 × 3) and all input channels (for example, number of feature maps). [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- Each 3D filter combines information across space and across all input feature maps to produce one new feature map. [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)



### Fully connected layer and softmax at the end

- After several rounds of convolution and pooling, the feature maps are small in height and width but rich in meaning. [towardsdatascience](https://towardsdatascience.com/convolutional-neural-networks-for-beginners-c1de55eee2b2/)

- These final feature maps are **flattened** into a 1D vector and passed to one or more **fully connected (dense) layers**. [adeshpande3.github](https://adeshpande3.github.io/A-Beginner's-Guide-To-Understanding-Convolutional-Neural-Networks/)

- The last layer is usually a **softmax** layer, which turns the final scores into probabilities over the possible classes (for example, cat, dog, car, etc.). [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- You can think of the convolution + pooling stack as automatically learning powerful features, and the final dense + softmax part as performing a logistic regression on those learned features. [adeshpande3.github](https://adeshpande3.github.io/A-Beginner's-Guide-To-Understanding-Convolutional-Neural-Networks/)



### What CNNs learn and why they need data

- CNNs **automatically learn** what features to detect; we do not tell them explicitly which shapes or patterns to look for. [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- The filters gradually specialize during training to capture useful patterns at different levels of abstraction. [en.wikipedia](https://en.wikipedia.org/wiki/Convolutional_neural_network)

- Learning these filters reliably usually requires large labeled datasets and significant computational power. [ibm](https://www.ibm.com/think/topics/convolutional-neural-networks)
