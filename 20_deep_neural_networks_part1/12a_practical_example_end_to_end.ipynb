```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from tensorflow.keras.utils import to_categorical
from sklearn.compose import make_column_transformer
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
df = load_wine(as_frame=True)

#Quick reminder on information contained in the data
print(df.DESCR)
```

```python
X, y = df.data, df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Preparing the data
# one hot encoded version of the *target*

y_ohe = to_categorical(y)
y_ohe[:5]
```

```python
tf.random.set_seed(42)
model1 = Sequential()
model1.add(Dense(50, activation = 'relu'))
model1.add(Dense(1, activation = 'sigmoid'))
model1.compile(loss = 'bce', metrics = ['acc'])
history1 = model1.fit(X_train, y_train, validation_data = (X_test, y_test),
                     epochs = 10, verbose = 0)
```

#### Regularization in the Network

Similar to what we have seen in early models, there are some options for regularization in your neural network.  Below, `model2` uses `kernel_regularizer = 'l1'` to apply L1 regularization to the hidden layer.  

```python
model2 = Sequential()
model2.add(Dense(50, activation = 'relu', kernel_regularizer = 'l1'))

model2.add(Dense(1, activation = 'sigmoid'))
model2.compile(loss = 'bce', metrics = ['acc'])
history2 = model2.fit(X_train, y_train, validation_data=(X_test, y_test),
           epochs = 10, verbose = 0)
```

#### Dropout

An alternative option for regularization is the `Dropout` layer.  This layer randomly "drops out" nodes in a given layer.

The Dropout layer randomly sets input units to 0 with a frequency of rate at each step during training time, which helps prevent overfitting. Inputs not set to 0 are scaled up by 1/(1 - rate) such that the sum over all inputs is unchanged.

```python
model3 = Sequential()
model3.add(Dense(50, activation = 'relu'))
model3.add(Dropout(0.2))
model3.add(Dense(1, activation = 'sigmoid'))
model3.compile(loss = 'bce', metrics = ['acc'])
history = model3.fit(X_train, y_train, validation_data=(X_test, y_test),
           epochs = 10, verbose = 0)
```
