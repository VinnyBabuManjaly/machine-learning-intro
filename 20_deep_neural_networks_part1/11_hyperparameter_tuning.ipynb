Hyperparameter tuning is the process of choosing the “settings” of a neural network (not the weights) so it learns well and generalizes to new data without overfitting.

#### What Are Hyperparameters?

Hyperparameters are design choices you set before training:

- Depth – number of layers (e.g., 1 hidden layer vs 3 hidden layers).  
- Width – number of neurons in each layer (e.g., 5 vs 32 per layer).  
- Activation functions – ReLU, tanh, sigmoid, softmax, etc.  
- Also: learning rate, batch size, number of epochs, etc. (mentioned elsewhere in the course).

These are not learned from the data like weights and biases; you choose them, and they strongly affect performance and training behavior.



#### Model Complexity and Parameters

As you add layers and neurons:

- The number of parameters (weights + biases) grows quickly.  
- For two fully connected layers with 32 neurons each, just the connections between them need $32 \times 32 = 1024$ weights, plus 32 biases.  
- With many such layers, you can end up with thousands of parameters even on a small dataset.

Effects of increasing complexity:

- Often faster convergence (loss drops quickly) because the model is very flexible.  
- But higher risk of overfitting: the model can fit noise or memorize the training data.

A simple model (few neurons and layers) has fewer parameters, is less flexible, and may underfit if the task is very complex, but it may generalize better on small datasets.



#### Using Validation (Dev) Sets and Cross-Validation

To choose good hyperparameters, you evaluate how well different architectures generalize.

Typical practice for neural networks:

- Use simple validation (train/dev split) rather than full k‑fold cross‑validation, because training is expensive.  
- Split the data into:
  - Training set – used to fit the model.  
  - Validation set (dev set) – held out to evaluate hyperparameters.

Process:

1. Shuffle the dataset.  
2. Split off a portion (e.g., 40 out of 150 samples) as validation.  
3. Train several candidate models (different depths/widths, etc.).  
4. For each model, track training loss and validation loss across epochs.  
5. Compare models by validation loss, not just training loss or accuracy.

The best architecture is usually the one with the lowest validation loss, not necessarily the most complex one.



#### Detecting Overfitting and Early Stopping

When you plot training and validation loss versus epochs:

- Training loss:
  - Typically keeps decreasing as the model fits the training data better.  
- Validation loss:
  - Decreases at first as the model learns useful patterns.  
  - Reaches a minimum.  
  - Then starts increasing as the model begins to overfit.

Key idea:  
- The epoch where validation loss is minimal is a good point to stop training (early stopping).  
- Training beyond this point usually means the model fits noise and generalizes worse.

Two models might have similar accuracy, but you may still prefer:

- The one with smoother, less noisy validation loss curves, suggesting more stable generalization.  
- Often this is the simpler model if both reach similar accuracy and validation loss.



#### Comparing Different Architectures

You can repeat this dev‑set process over many architectures:

- Vary depth (number of layers).  
- Vary width (number of neurons per layer).  
- Record validation loss for each configuration.

Then:

- Build a table or plots (e.g., validation loss vs depth, or vs width).  
- Choose the architecture with lowest validation loss, subject to simplicity and stability.

Remember:

- The validation loss you see is a bit optimistic because you used it to choose the model.  
- For a final, unbiased performance estimate, you should evaluate once on a separate test set that you do not touch during tuning.



#### Automated Hyperparameter Search

Instead of manually trying many combinations, specialized tools can:

- Search over a predefined set of hyperparameters (e.g., different numbers of layers, neurons, learning rates).  
- Run training for each combination and compare validation losses.  

These tools behave similarly to grid search methods used for simpler models, but tailored for neural networks.



##### Core Concepts Summary

- Hyperparameters define the architecture and training configuration of a neural network (depth, width, activations, etc.).  
- Model complexity increases with more layers/neurons and can both help and hurt.  
- Validation sets and validation loss curves are central to picking sensible hyperparameters and avoiding overfitting.  
- You typically choose the simplest model that achieves low, stable validation loss, and you use a separate test set for final evaluation.

### What Are Hyperparameters?

Hyperparameters are settings you choose before training a model. They:

- Stay fixed during training.
- Strongly influence how the model learns and how complex it is.
- Are not learned from data (unlike weights and biases).

Two broad types:

1. Model hyperparameters – define the model’s structure  
   - Example for neural networks: number of hidden layers, number of neurons in each layer (width).

2. Algorithm hyperparameters – define how the learning algorithm behaves  
   - Example: learning rate for gradient descent, number of neighbors in k‑NN.

Choosing good hyperparameters is called hyperparameter tuning. Tools like KerasTuner automate this, similar to grid search or random search in traditional machine learning.



### Dropout Rate

Dropout is a regularization technique used to prevent overfitting in neural networks.

- During training, some neurons are randomly “dropped out” (ignored) on each step.  
- Dropped neurons:
  - Do not contribute to forward propagation (their outputs are treated as zero).
  - Do not get their weights updated in backpropagation.

The dropout rate is the probability of dropping a neuron (e.g., 0.5 means drop 50% of neurons in that layer each step).

Intuition:

- It forces the network to not rely too heavily on any single neuron or pattern.  
- Encourages redundancy and more robust, generalizable features.



### Regularization Rate

Regularization is a family of techniques to reduce model complexity and combat overfitting by adding a penalty on large weights.

Two common forms:

#### L1 Regularization

- Adds a penalty proportional to the absolute values of the weights.  
- Encourages many weights to become exactly zero.  
- Produces sparse models (few nonzero coefficients).  
- Often associated with Lasso in regression.

Effect: the model keeps only the most important connections and can “turn off” others.

#### L2 Regularization

- Adds a penalty proportional to the square of the weights.  
- Shrinks all weights toward zero, but usually none become exactly zero.  
- Keeps all features but reduces their influence.  
- Used in many models, including regression and SVMs.

Effect: the model still uses all connections but with smaller, more conservative weights, which helps reduce overfitting.

The regularization rate (often called λ or alpha) controls how strong this penalty is. Higher rates mean more aggressive shrinking.



### Big Picture

- Hyperparameters define how the model is built and trained.  
- Dropout randomly disables neurons during training to improve generalization.  
- L1/L2 regularization penalize large weights to simplify the model and reduce overfitting.  
- Tools like KerasTuner help automatically search for good combinations of these hyperparameters.
